```python
from seaborn import heatmap
from task import draw_plots, draw_plot
import pandas as pd
import json as js
import matplotlib.pyplot as plt
from seaborn import heatmap
```

```python
with open('data.json') as file:
    data = pd.DataFrame(js.load(file), columns=
        [
            'gt_corners', 
            'rb_corners',
            'mean',
            'max',
            'min',
            'floor_mean',
            'floor_max',
            'floor_min',
            'ceiling_mean',
            'ceiling_max',
            'ceiling_min',
        ]
    )

    # Shows correlation of dataFrame
    heatmap(data.corr(), annot=True, fmt='.2f')
    plt.show()
    plt.close()

    # Shows gt_corners vs rb_corners(There is an accuracy of the model)
    plt.xlabel('gt_corners', color='red')
    plt.ylabel('rb_corners', color='red')
    plt.grid()
    plt.plot(data.gt_corners, data.rb_corners, marker='o', markersize='2', linestyle='None')
    plt.show()
    plt.close()

    # Shows floor_max vs ceiling_max
    plt.xlabel('floor_max', color='red')
    plt.ylabel('ceiling_max', color='red')
    plt.grid()
    plt.plot(data.floor_max, data.ceiling_max, marker='o', markersize='2', linestyle='None')
    plt.show()
    plt.close()

    # Shows floor_min vs ceiling_min
    plt.xlabel('floor_min', color='red')
    plt.ylabel('ceiling_min', color='red')
    plt.grid()
    plt.plot(data.floor_min, data.ceiling_min, marker='o', markersize='2', linestyle='None')
    plt.show()
    plt.close()

    # Shows floor_mean vs ceiling_mean
    plt.xlabel('floor_mean', color='red')
    plt.ylabel('ceiling_mean', color='red')
    plt.grid()
    plt.plot(data.floor_mean, data.ceiling_mean, marker='o', markersize='2', linestyle='None')
    plt.show()
    plt.close()
```
